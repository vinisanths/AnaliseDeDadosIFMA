```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = '/content/sales.csv'
df = pd.read_csv(data)
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
#plotar grafico
df['Sales'].plot(kind='line', figsize=(8,4), title='Sales')
plt.show()
```

```python
# ordenar em ordem decrescente as variveis por seus valores ausentes
(df.isnull().sum() / df.shape[0]).sort_values(ascending=False)
```
