#Criar 1º Dataframe

```python
nomes = ['Jose', 'Maria','Luis','Fernanda','Lucio', 'Raquel']
idades = [21,20,19,18,35,32]
dados = list(zip(nomes,idades))
print(dados)
```

```python
#Criando e exibindo Dataframe
df1 = pd.DataFrame(data=dados, columns=['Nome','Idade'])
df1
```

```python
#Criar Dataframes via Criação de rotulos
dados1 = [('Amalia',24), ('Joao', 22),('Maria', 27),('Fernanda',26)]
colunas = ['Nome','Idade']
linhas = ['a','b','c','d']

df2 = pd.DataFrame(data=dados1, columns=colunas, index=linhas)
df2
```

```python
#Criar Dataframes via Dicionarios
dados2 = {'nome': ['Ana','Maria','Joana'],
         'idade': [20,30,38]}

df3 = pd.DataFrame(data=dados2)
df3
```
